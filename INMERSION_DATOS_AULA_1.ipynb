### MACHINE LEARNIG MODEL

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x= datos_ml[['COD_UPZ_GRUPO']]
y= datos_ml[['Precio_Millon']]
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.25, random_state=99)
```

```python
#Linear regression model
modelo=LinearRegression()
```

```python
# Model training
modelo.fit(x_train, y_train)
```

```python
# Assign dependent and independent variables
y_predict_test= modelo.predict(x_test)
```

```python
# Model validation
# Medium error and  coeficient of determination of the model, as we can see, it has a huge error and a big dispersion among the dependent and independent variable, which in this case is the price
from sklearn.metrics import mean_absolute_error, r2_score
baseline_mae= mean_absolute_error(y_test, y_predict_test)
baseline_r2= r2_score(y_test, y_predict_test)
print(baseline_mae, baseline_r2)
```
